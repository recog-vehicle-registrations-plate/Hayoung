```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract

plt.style.use('dark_background')
```

```python
conf = ('--tessdata-dir "/usr/share/tesseract-ocr/4.00/tessdata" --oem 2 --psm 7')
#conf = ('--tessdata-dir "/usr/share/tesseract-ocr/4.00/tessdata" --oem 2 --psm 13') ==> 별로다

```

```python
#img_ori = cv2.imread('/home/pirl/Documents/ESRGAN/results/RJ.png')
img_ori=cv2.imread('02282234414511492.jpg')
height, width, channel = img_ori.shape

plt.figure(figsize=(10,12))
plt.imshow(img_ori, cmap='gray')
print(pytesseract.image_to_string(img_ori, lang='eng', config=conf))
```

```python
gray = cv2.cvtColor(img_ori,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,12))
plt.imshow(gray, cmap='gray')
print(pytesseract.image_to_string(gray, lang='eng', config=conf))
```

```python

hist1 = cv2.calcHist([img_ori],[0],None,[256],[0,256])

plt.subplot(221),plt.imshow(img_ori,'gray'),plt.title('Red Line')
plt.subplot(223),plt.plot(hist1,color='r')
plt.xlim([0,256])
plt.show()
```

```python
img_blurred =cv2.GaussianBlur(gray, ksize=(3,3), sigmaX=0)

img_thresh = cv2.adaptiveThreshold(
    img_blurred,
    maxValue = 255.0,
    adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    thresholdType = cv2.THRESH_BINARY_INV,
    blockSize=101, # originally, blockSize = 19
    C=-5)

plt.figure(figsize=(width,height))
plt.imshow(img_thresh, cmap='gray')
print(pytesseract.image_to_string(img_thresh, lang='eng', config=conf))
```

```python
contours, _ = cv2.findCon식tours(
    img_thresh, 
    mode=cv2.RETR_LIST, 
    method=cv2.CHAIN_APPROX_SIMPLE
)

temp_result = np.zeros((height, width, channel), dtype=np.uint8)

cv2.drawContours(temp_result, contours=contours, contourIdx=-1, color=(255, 255, 255))

plt.figure(figsize=(2,3))
plt.imshow(temp_result)
print(pytesseract.image_to_string(temp_result, lang='eng', config=conf))
```
